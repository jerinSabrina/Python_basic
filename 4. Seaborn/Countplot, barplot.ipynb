```python
import seaborn as sns
```

```python
df = sns.load_dataset("iris")
df.head()
```

```python
sns.countplot(x ='species',data = df)
```

```python
#Cannot pass values for both `x` and `y`
#either x or y
sns.countplot(x= 'petal_width',data = df)
```

```python
sns.countplot(x = 'sepal_length',y = None, palette='twilight',data =df)
```

```python
sns.countplot(x = 'sepal_length',y = None,hue = 'species', palette='rainbow', data =df)
```

```python
sns.countplot('petal_width',data=df,orient = 'v')
```

```python
sns.countplot(x = 'sepal_length',y = None,hue = 'species', palette='rainbow', dodge=False, data =df)
```

```python
sns.countplot(x = 'petal_width',y = None,hue = 'species', palette='BuPu_r', saturation=1, data =df)
```
