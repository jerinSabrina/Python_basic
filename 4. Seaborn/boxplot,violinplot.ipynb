```python
import seaborn as sns
df = sns.load_dataset("iris")
df.head()
```

```python
sns.boxplot('sepal_length','species',data=df)
```

```python
sns.boxplot('species','sepal_length',data=df)
```

```python
sns.boxplot(data=df,orient = 'v')
```

```python
sns.boxplot(data=df,orient = 'h')
```

```python
sns.boxplot(x = 'sepal_width',y ='petal_width',data=df,hue = 'species')
```
