# Matplotlibrary

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0,10)
y = x*x
plt.scatter(x,y,c='b')
plt.xlabel("X axis")
plt.ylabel('Y axis')
plt.title('Graph plotting')
plt.savefig('fig.jpg')
```

# Plot Function

```python
plt.plot(x, y, color='g', marker='o', linestyle='dashed',linewidth=2, markersize=15)
```

# Subplot

```python
a =np.arange(0,5)
b = np.sqrt(a)
plt.subplot(2,1,1)
plt.plot(a,b,'r--')
plt.subplot(2,1,2)
plt.plot(a,b,'b*-')
```

```python
# y=a*x^2 + b*x + c funtion
x= np.arange(10,20,step = 2)
y = 3 * x^2 + 5*x +5
y1 = 3*x +5
plt.subplot(1,2,1)
plt.plot(x,y,'^k:')
plt.subplot(1,2,2)
plt.plot(x,y1,'^g:')
```

# Bar Plot

```python
x =[2,5,10]
y= [11,20,5]
x2 = [3,9,11]
y2= [6,15,7]
plt.bar(x,y,color= 'g')
plt.bar(x2,y2,color='r')
plt.title('Bar graph')
```

# Histogram

```python
#a =np.array([22,20,17,4,5,12,52,35,25,14,3,8,10])
#a=np.sort(a)
a1 = np.array([4,5,8,10,12,14,15,17,20,22,25,35,42,42])
plt.hist(a1,bins =10,range = None,cumulative = False)
plt.title('Histogram')
```

# Pie Chart plot

```python
s= [25,40,15,20]
exp = (0,0,0.2,0)
labels = "shimu",'sajid','sharmin','jerin'
c = ['green','yellow','red','blue']
plt.pie(s,explode=exp,labels=labels,colors= c,autopct='%1.01f%%',pctdistance=0.6,shadow=True)
plt.title('PieChart')
```
